## Pivot Table:  Number of Counties (by State) that have Turbines from each Manufacturer

```python
import pandas as pd
# Authored in anaconda.cloud account
print(pd.__version__)
```

```python
url = 'https://gist.githubusercontent.com/mattkram/5078f1cb198694274cb708253b0dc210/raw/f90167d071abe328f7f437f3d68dbd5deb6012c8/uswtdb_v5_3_20230113.csv'
df = pd.read_csv(url)
df.info()
```

```python
columns_to_keep = ['p_year', 't_state', 't_county', 't_model', 't_manu', 'faa_ors']
df = df[columns_to_keep]
pd.set_option("display.max_rows", 16)
```

```python
df = df[df['t_state'].isin(['DE', 'MN', 'MA', 'OR'])]
```

### Now that rows & columns have been reduced, drop any records with NaN and sort by year 

```python
sliced = df.dropna()
```

```python
diced = sliced.sort_values(by='p_year')
diced
```

```python
grouped = diced.groupby(['t_manu', 't_state'])
```

### Count of counties that have each type of wind turbine

```python
counted = grouped['t_county'].nunique().reset_index()
counted
```

```python
counted = counted[~counted['t_manu'].str.contains('changzhou', case=False)]
```

### Pivot table by State:  Number of counties with each type

```python
pd.set_option("display.max_rows", 40)
pivoted = counted.pivot_table(index='t_manu', columns='t_state', values='t_county').fillna(0).astype('int64')
pivoted
```
